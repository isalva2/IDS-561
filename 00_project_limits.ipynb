```python
import geopandas as gpd
import matplotlib.pyplot as plt

shapefile_path = "data/spatial/geo_export_86dc231e-3ba9-473c-a7a5-0e89f969d1f6.shp"
gdf = gpd.read_file(shapefile_path)

# Create a figure and axis object to control plot size
fig, ax = plt.subplots(figsize=(10, 8), dpi=200)  # Adjust the figsize as needed (width, height)

# Plot the GeoDataFrame
gdf.plot(ax=ax)

# Add labels for each shape in the GeoDataFrame
for idx, row in gdf.iterrows():
    ax.text(row.geometry.centroid.x, row.geometry.centroid.y, row['zip'], ha='center', fontsize=2)

plt.show()
```
